```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
city=pd.read_csv(r"C:/Users/Dell/OneDrive - Arab Open University - AOU/Desktop/M238/TMA_Spring_23-24-20240410/Cities.csv")
continent=pd.read_csv(r"C:/Users/Dell/OneDrive - Arab Open University - AOU/Desktop/M238/TMA_Spring_23-24-20240410/continent.csv")
world_cities=pd.read_csv(r"C:/Users/Dell/OneDrive - Arab Open University - AOU/Desktop/M238/TMA_Spring_23-24-20240410/worldcities.csv")
```

## First let's merge all the three datasets by using pd.merge() function.

```python
merged_df = pd.merge(city, world_cities, left_index=True, right_index= True)
merged_df
```

```python
merged_df=pd.merge(merged_df, continent, left_index=True, right_index=True)
merged_df[:5]
```

## after merging the three datasets let's remove the columns that we won't be using 

```python
df=merged_df.drop(columns=["Id","SmartCity_Index", "admin_name","capital","id","Country_Name","Two_Letter_Country_Code","Three_Letter_Country_Code","Country_Number","SmartCity_Index_relative_Edmonton","city","city_ascii","country","iso2","iso3"])
# let's see the first five cities with the columns we need .
df[:5]
```

# Here i am going to do some preprocessing of the dataset.

## first let's check if there is some empty rows 

```python
df.isnull().sum()
```

```python
# in the above code we can see that in the Continent_Code we have 18 Na values whic resembles as North America so python recognizes them as Na values so let's change them
df.fillna('North AM', inplace=True)
```

```python
# here after replacing the Na in continent code with North Am there is no more Null values.
df.isnull().sum()
```

## Visualizing ten cities by using heatmap .

### we can use correlation between columns to present the data as a heatmap

```python
# Select the chosen cities to appear in our visualization.
selected_cities=["Dubai","Abu Dhabi","Beijing","Tokyo","Osaka","Seoul","Moscow","Singapore","Shanghai","Taipei"]

# making sure that we only choose the selected cities from the city column in our dataset.

selected_df = df[df['City'].isin(selected_cities)]
selected_df.set_index('City', inplace=True)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(selected_df[['Smart_Mobility', 'Smart_Environment', 'Smart_Government', 'Smart_Economy', 'Smart_People', 'Smart_Living']], cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.ylabel('City')
plt.show()
```

```python
import folium

# List of city coordinates
city_coordinates = [
    {"city": "Dubai", "latitude": 25.276987, "longitude": 55.296249},
    {"city": "Abu Dhabi", "latitude": 24.4539, "longitude": 54.3773},
    {"city": "Tokyo", "latitude": 35.6895, "longitude": 139.6917},
    {"city": "Osaka", "latitude": 34.6937, "longitude": 135.5023},
    {"city": "Shanghai", "latitude": 31.2304, "longitude": 121.4737},
    {"city": "Beijing", "latitude": 39.9042, "longitude": 116.4074},
    {"city": "Seoul", "latitude": 37.5665, "longitude": 126.9780},
    {"city": "Taipei", "latitude": 25.0330, "longitude": 121.5654},
    {"city": "Singapore", "latitude": 1.3521, "longitude": 103.8198},
    {"city": "Moscow", "latitude": 55.7558, "longitude": 37.6176}
]

# Create a map centered around Dubai
map_center = [25.276987, 55.296249]
map_show = folium.Map(location=map_center, zoom_start=5)

# Add markers for each city in city_coordinates
for city in city_coordinates:
    folium.CircleMarker(
        location=[city['latitude'], city['longitude']],
        radius=30,  # Specify the radius of the circle marker
        color='dodgerblue',  # Outline color of the circle marker
        fill=True,
        fill_color='dodgerblue',  # Fill color of the circle marker
        fill_opacity=0.7,
        popup=city['city']
    ).add_to(map_show)

# Display the map
map_show
```
